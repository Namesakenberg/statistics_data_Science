```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_samples=1000 # total size of the sample
sample_size=30  # maximum number of values that can be taken at a time in a sample (maximum values in a set)
sample=np.random.uniform(low=0.0, high=1.0,size=(num_samples,sample_size))
```

```python
sns.kdeplot(sample)
```

```python
sample_mean =np.mean(sample,axis=1)
sample_mean

# while calculating the mean :
# axis =0 : gives mean of the each feature or column
# axis =1 : gives mean for every sample
# axis = none = gives mean of the entire data
```

```python
# plot the mean
sns.kdeplot(sample_mean)
```

# we can see that , though the original data was uniformly distributed , the distribution of sample mean of the sample distriution will always be normally distributed

```python
# if the mean of the  sample distribution is mu and the variance is sigma^2
# then
# the mean of the normal distribution formed will be mu and is variance will be sigma^2 / n
```

```python
mean_sample = np.mean(sample)
variance_sample = np.var(sample)
print("the mean of the sample is ",mean_sample)
print("the variance of the sample is : ",variance_sample)
```

```python
mean_normdist = np.mean(sample_mean)

variance_normdist = np.var(sample_mean)

print("the mean of the sample is ",mean_normdist)
print("the variance of the sample is : ",variance_normdist)

check = variance_sample/sample_size
print(check)
```

```python
df = df.dropna(subset=['fare'])  # Drops rows where 'fare' has NaN values
```

```python
df['fare']
```

```python
samples=[] # 2d list to store points
for i in range(100):
  samples.append(df['fare'].sample(50).values.tolist())
```

```python
samples=np.array(samples)
```

```python
# find mean of a particular sample
sample_mean=np.mean(samples,axis=1)
```

```python
sample_mean.shape
```

```python
# plot
sns.kdeplot(sample_mean)
```

```python
# the graph obtained is not exactly similar to normal distribution but is similar to it
```

```python
sample_mean.mean()
```

```python
sample_mean.std()
```

```python
stddev = sample_mean.std()/7.0710678118654755
```

```python
lower_val = sample_mean.mean()-2*(stddev)
```

```python
higher_val = sample_mean.mean()+2*(stddev)
```
