# Using Z procedure

```python
# the std of the fare column is 51.758
# make samples
# find sample mean of sample distribution
# consider the confidence interval of 95 percent
# use the formula ci = X +- Za/2*sigma/root n

n=1310
sigma= 51.758
# the population is 1310 we draw 50 samples as there should atleast be 30 samples
# calculate the sample mean of sample distribution
```

```python
# shuffle the data to ensure randomness
df.sample(1308)
```

```python
df['fare']
```

```python
samples=[]
for i in range(100):
  samples.append(df['fare'].sample(30).values.tolist())
```

```python
samples=np.array(samples)
```

```python
sample_mean=np.mean(samples,axis=1)
```

```python
sample_mean.shape
```

```python
sample_mean
```

```python
sample_dev=sample_mean.std()
```

```python
# get the mean of sample_mean
pointestimate = sample_mean.mean()
```

```python
print('the point estimate value is ',pointestimate)
```

```python
# instead of caluclating sample mean of sample distribution , if we find the sample mean its still fine
```

```python
df['fare'].mean()
```

```python
#Confidence interval :
# considering then confidence level as 95 percent(0.95) then the value of upper bound would be (0.95+0.025=0.0975)
# value of 0.0975 in the Z table is 1.96
# hence lower bound's point = -1.96 and upper bound's point is 1.96
upper_bound = pointestimate + ( 1.96 * ( sigma / n**0.5 ) )
lower_bound = pointestimate - ( 1.96 * ( sigma / n**0.5 ) )
```

```python
print(f'hence the confidence interval is {lower_bound} {upper_bound} with confidence value of 95%')
```

```python
#Confidence interval :
# considering then confidence level as 99 percent(0.99) then the value of upper bound would be (0.99+0.005=0.0995)
# value of 0.0995 in the Z table is 2.58
# hence lower bound's point = -2.58 and upper bound's point is 2.58
upper_bound = pointestimate + ( 2.58 * ( sigma / n**0.5 ) )
lower_bound = pointestimate - ( 2.58 * ( sigma / n**0.5 ) )
```

```python
print(f'hence the confidence interval is {lower_bound} {upper_bound} with confidence value of 99%')
```

hence with increasing confidence level the margin of error also increases

 # using T procedure


```python
# cI = sample_mean +- Ta/2 * (S/root(n))
lower_T = pointestimate - ((2.042)*(sample_dev/n**0.5))
lower_T
```

```python
upper_T = pointestimate + ((2.042)*(sample_dev/n**0.5))
```

```python
print(f'hence the confidence interval is {lower_T} {upper_T} with confidence value of 95%')
```
